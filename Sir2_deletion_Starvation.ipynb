```python
from draft0 import *
import os
import time
import rpy2.robjects as R
import matplotlib.patches
%matplotlib inline
```

```python
dfs=[]
for data_file in ['/Users/q6600sl/Documents/Mito_Sir2_Starvation/Sir2_st1_mito_mother.xls',
                  '/Users/q6600sl/Documents/Mito_Sir2_Starvation/Sir2_st1_w1118_mother.xls',
                  '/Users/q6600sl/Documents/Mito_Sir2_Starvation/Sir2_st2_mito_mother.xls',
                  '/Users/q6600sl/Documents/Mito_Sir2_Starvation/Sir2_st2_w1118_mother.xls']:
    _df = pd.read_excel(data_file)
    _df = _df.T.fillna(method='pad').diff().T
    _df.columns = _df.columns.map(lambda x: time.mktime(time.strptime(str(x), '%Y%m%d%H%M')))/3600
    _df.columns = _df.columns - _df.columns.min()
    _df.fillna(0, inplace=True)
    _fd = _df.copy()
    _fd.columns = np.hstack([0, _df.columns[:-1]])
    _df = pd.DataFrame(_df.stack())
    _df.reset_index(inplace=True)
    _df.columns = ['mito', 'Sir2', 'Food', 'vail_pos', 'Time1', 'Count']
    _fd = pd.DataFrame(_fd.stack())
    _fd.reset_index(inplace=True)
    _fd.columns = ['mito', 'Sir2', 'Food', 'vail_pos', 'Time0', 'Count']
    _df['Time0'] = _fd['Time0'].values
    _df = _df[_df.Count>0]
    _df = pd.merge(_df,
                   pd.DataFrame({'Val': np.arange(_df.Count.sum())+1}),
                   left_on = _df.Count.cumsum(),
                   right_on = 'Val',
                   how = 'outer').sort('Val').fillna(method = 'bfill').reset_index(drop=True)
    _df['Rep'] = int(data_file.split('_')[-3][-1])
    _df['Mother'] = data_file.split('_')[-2]
    _df = _df[['mito', 'Sir2', 'Food', 'Rep', 'Mother', 'Time0', 'Time1']]
    dfs.append(_df)
```

```python
df = pd.concat(dfs).reset_index(drop=True)
df.to_csv('/Users/q6600sl/Documents/Mito_Sir2_Starvation/Sir2_st_data.csv',
          index=None)
```

```python
df_gby  = df.groupby(['mito', 'Sir2', 'Food', 'Rep', 'Mother']).agg(lambda x: tuple(x))
prm_est = [np.hstack(fit_dist('lg2', df_gby.ix[i,0], df_gby.ix[i,1], step=.1)) for
           i in range(df_gby.shape[0])]
df_prm = pd.DataFrame(np.vstack(prm_est))
df_prm.columns = ['log_u','log_s','logLik', 'optimizer_flag']
df_prm.index = df_gby.index
df_prm.head()
```
