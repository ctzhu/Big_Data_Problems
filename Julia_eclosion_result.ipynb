```python
import os
import pandas.rpy.common as com
import pandas as pd
import rpy2.robjects as ro

os.chdir('Julia_Donner_data')
```

```python
df1 = pd.read_csv('eggpick1_csedata.csv').icol([0,2,3,4,6,7,8,10,11,12,14,15,16])
df1.set_index('hours', drop=True, inplace=True)
df1.columns = pd.MultiIndex.from_product((('19B','19G','15B','15G'),('cse0','cse50','cse100')),
                                         names=['Genotype','CSE_level'])
df1
```

```python
df2 = pd.read_csv('eggpick2_csedata.csv').icol([0,2,3,4,6,7,8,10,11,12,14,15,16])
df2.set_index('hours', drop=True, inplace=True)
df2.columns = pd.MultiIndex.from_product((('19B','19G','15B','15G'),('cse0','cse50','cse100')),
                                         names=['Genotype','CSE_level'])
df2
```

```python
df3 = pd.read_csv('eggpick3_csedata.csv').icol([0,2,3,4,6,7,8,10,11,12,14,15,16])
df3.set_index('hours', drop=True, inplace=True)
df3.columns = pd.MultiIndex.from_product((('19B','19G','15B','15G'),('cse0','cse50','cse100')),
                                         names=['Genotype','CSE_level'])
df3
```

```python
pd_all = pd.concat((pd.DataFrame({'Count': df1.stack([0,1]), 'Rep':1}),
                    pd.DataFrame({'Count': df2.stack([0,1]), 'Rep':2}),
                    pd.DataFrame({'Count': df3.stack([0,1]), 'Rep':3})), axis=0).reset_index()
pd_all['total_N'] = 60
pd_all['Nuc']     = 'RAL'+pd_all.Genotype.str.extract('([0-9]+)')
pd_all['Mito']    = pd_all.Genotype.str.extract('([A-Z]+)')

pd_all.head()
```

```python
pd.merge(pd.DataFrame({'hours': sorted(pd_all.hours.unique()),
                       'time_bin': range(1, 1+len(pd_all.hours.unique()))}),
         pd_all,
         how='right',
         on='hours').to_csv('JD_cse_eggpick.csv', index=None)
```

```python
ro.reval("egg_pick <- read.csv('/Users/q6600sl/IPython_NB/Julia_Donner_data/JD_cse_eggpick.csv')")

#third-order orthogonal polynomial

ro.reval("t <- poly(unique(egg_pick$time_bin), 3)")
ro.reval("egg_pick[, paste('ot', 1:3, sep='')] <- t[egg_pick$time_bin, 1:3]")
```

```python
ro.reval('library(lme4)')

#logistic GCA

r_code = '''m_log <- glmer(cbind(Count, total_N-Count) ~ (ot1+ot2+ot3)*Nuc +
                                             (ot1+ot2+ot3)*Mito+
                                             (ot1+ot2+ot3)*CSE_level+
                                             (ot1+ot2+ot3|Rep), 
               data=egg_pick, family=binomial)'''
ro.reval(r_code)
ro.reval('smy <- summary(m_log)')
```
