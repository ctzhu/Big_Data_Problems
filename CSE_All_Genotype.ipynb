```python
def walk_folder(start_folder):
    folder = os.listdir(start_folder)
    folder = [os.path.join(start_folder, item) for item in folder]
    for item in folder:
        if (not isinstance(item, list)) and os.path.isdir(item):
            folder.extend(walk_folder(item))
    return list(set(folder))
```
